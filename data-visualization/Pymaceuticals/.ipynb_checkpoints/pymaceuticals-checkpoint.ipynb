```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combined_data = pd.merge(study_results, mouse_metadata, how="left")
# Display the data table for preview
combined_data.head()
```

```python
# Checking the number of mice.
mice_count = mouse_metadata["Mouse ID"].count()
mice_count
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = combined_data.loc[combined_data[['Mouse ID', 'Timepoint']].duplicated(), 'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_df = combined_data.loc[combined_data['Mouse ID'] == 'g989']
duplicate_mouse_df.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
new_df = combined_data[combined_data['Mouse ID'].isin(duplicate_mice)==False]
new_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
df_mice_count = new_df["Mouse ID"].nunique()
df_mice_count
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
drug_regimen_df = new_df.groupby('Drug Regimen')
# mean, median, variance, standard deviation, and SEM of the tumor volume.
drug_regimen_mean = drug_regimen_df['Tumor Volume (mm3)'].mean()
drug_regimen_median = drug_regimen_df['Tumor Volume (mm3)'].median()
drug_regimen_variance = drug_regimen_df['Tumor Volume (mm3)'].var()
drug_regimen_std = drug_regimen_df['Tumor Volume (mm3)'].std()
drug_regimen_sem = drug_regimen_df['Tumor Volume (mm3)'].sem()
# Assemble the resulting series into a single summary DataFrame.
drug_regimen_summary = pd.DataFrame({"Mean Tumor Volume":drug_regimen_mean,
                                   "Median Tumor Volume":drug_regimen_median,
                                   "Tumor Volume Variance":drug_regimen_variance,
                                   "Tumor Volume Std. Dev.":drug_regimen_std,
                                   "Tumor Volume Std. Err.":drug_regimen_sem})
drug_regimen_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.
drug_regimen_summary_a = drug_regimen_df['Tumor Volume (mm3)'].aggregate(["mean", "median","var","std","sem"])
drug_regimen_summary_a
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
Capomulin_final = new_df.loc[new_df["Drug Regimen"]=="Capomulin", :]
Ramicane_final = new_df.loc[new_df["Drug Regimen"]=="Ramicane", :]
Infubinol_final = new_df.loc[new_df["Drug Regimen"]=="Infubinol", :]
Ceftamin_final = new_df.loc[new_df["Drug Regimen"]=="Ceftamin", :]

#print(Capomulin_final)

# Start by getting the last (greatest) timepoint for each mouse

last_timepoint = new_df.groupby(['Mouse ID'])['Timepoint'].max()
last_timepoint_reset = last_timepoint.reset_index()
last_timepoint_df = pd.DataFrame(last_timepoint_reset)

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint

merged_last_timepoint = pd.merge(last_timepoint_df, combined_data, on=("Mouse ID","Timepoint"),how="left")
merged_last_timepoint
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers.

def outlier(treatments):
    tumor_volume = merged_last_timepoint.loc[merged_last_timepoint["Drug Regimen"]==treatments]['Tumor Volume (mm3)']
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    quartiles = tumor_volume.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    iqr_1_5 = iqr * 1.5

#   tumor_volume = regimen["Tumor Volume (mm3)"]

#     print(f"The lower quartile of Tumor Volume is: {lowerq}")
#     print(f"The upper quartile of Tumor Volume is: {upperq}")
#     print(f"The interquartile range of Tumor Volume is: {iqr}")
#     print(f"The median of Tumor Volume is: {quartiles[0.5]}")
    
# Determine outliers using upper and lower bounds
    lower_bound = lowerq - iqr_1_5
    upper_bound = upperq + iqr_1_5
    
    # add subset 
    subset = tumor_volume.loc[(tumor_volume < lower_bound) | (tumor_volume > upper_bound)]
    print(f"{treatments} potential outliers:{subset}")
    return tumor_volume
    
#     print(f"Values below {lower_bound} could be outliers.")
#     print(f"Values above {upper_bound} could be outliers.")

Capomulin = outlier("Capomulin")
Ramicane = outlier("Ramicane")
Infubinol = outlier("Infubinol")
Ceftamin = outlier("Ceftamin")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
dot = dict(markerfacecolor='red')
plt.boxplot([Capomulin, Ramicane, Infubinol, Ceftamin], labels=treatments, flierprops=dot)
plt.ylabel("Final Tumor Volume (mm3)")
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

Capomulin_df = new_df.loc[new_df["Drug Regimen"]=="Capomulin", :]
id_I509 = Capomulin_df.loc[Capomulin_df["Mouse ID"]=='l509', :]

Capomulin_tumor_volume = id_I509["Tumor Volume (mm3)"]
Capomulin_timepoint = id_I509["Timepoint"]

plt.plot(Capomulin_timepoint, Capomulin_tumor_volume)

plt.title("Capomulin treatment of mouse l509")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint (days)")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

Capomulin_regimen = Capomulin_df.groupby(["Mouse ID"]).mean()
mouse_weight = Capomulin_regimen["Weight (g)"]
Capomulin_avg_tumor = Capomulin_regimen["Tumor Volume (mm3)"]

plt.scatter(mouse_weight, Capomulin_avg_tumor)
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
correlation = st.pearsonr(mouse_weight,Capomulin_avg_tumor)
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

(slope, intercept,rvalue, pvalue, stderr) = linregress(mouse_weight, Capomulin_avg_tumor)
regress_values = mouse_weight * slope + intercept
plt.scatter(mouse_weight,Capomulin_avg_tumor)
plt.plot(mouse_weight, regress_values, "r-")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```
